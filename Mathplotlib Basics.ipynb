# import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Basic Graphs


```python
plt.plot([1,2,3],[4,5,6])
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)

# adding a title to the graph
plt.title('line graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})

# adding labels to the axes
plt.xlabel('X axis',fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('y axis',fontdict={'fontname':'Comic Sans MS'})

plt.show()
```

# changing tick marks

```python
# changing tick marks
plt.xticks([0,1,2,3,4,10])
plt.yticks([0,2,4,6,8,10])
plt.plot(x,y,label='zxy')

plt.show()
```

# add a legend

```python
plt.plot(x,y,label='2x')
```

# Restyling the line

```python
plt.plot(x,y,color='red',linewidth=1.5,linestyle='--',marker='.',markersize='15',markeredgecolor='blue') 
# hex codes of colors can also be passed
```

```python
plt.plot(x,y,label='2x')
x2=np.arange(0,4,0.5)
print(x2)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[:4],x2[4:]**2,'r--')
```
